# SHAP + LIME explanations (separate step)

This notebook runs explanations **after** models exist under `Results/`.

This can be slow on a CPU simulator. Use conservative caps.

## Run SHAP + LIME on all stored models

This loads trained models and replays SHAP/LIME against the stored train/test splits.

It writes into:
- `Results/<dataset>/<method>/explain/shap/`
- `Results/<dataset>/<method>/explain/lime/`

```
!python examples/run_explain_all.py \
  --results-dir Results \
  --max-features 16 \
  --max-test-samples 2 \
  --background-samples 8 \
  --shap-max-evals 256 \
  --lime-num-samples 300
```
